```python
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
gc = pd.read_csv("C:/Users/Andrew Sundberg/Documents/DS301/german_credit_data.csv")
```

```python
gc.head()
```

```python
df = gc.dropna()
```

```python
#df_1 = pd.concat([df,pd.get_dummies(df['Sex'],prefix = 'Sex'),pd.get_dummies(df['Housing'],prefix = 'Housing'), pd.get_dummies(df['Duration'],prefix = 'Duration'), pd.get_dummies(df['Saving accounts'],prefix = 'Saving accounts'), pd.get_dummies(df['Checking account'],prefix = 'Checking account'), pd.get_dummies(df['Purpose'],prefix = 'Purpose')])
categorical = ['Sex','Housing','Saving accounts', 'Checking account', 'Purpose']
df_1 = pd.get_dummies(df, prefix_sep = '__', columns = categorical)
df_1.head()
```

```python
X = df_1.loc[:, df_1.columns != 'Risk']
y = df_1.loc[:, df_1.columns == 'Risk']
print("X.shape=", X.shape)
print("y.shape=", y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state = 100)
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(solver = 'lbfgs')
logreg.fit(X_train,y_train)
logreg.predict(X_test)
y_pred = logreg.predict(X_test)
y_true = y_test
```

```python
len(y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_true, y_pred)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegressionCV
clf = LogisticRegressionCV(cv = 5, random_state = 0, multi_class = 'multinomial').fit(X,y)
clf.predict(X_test)
clf.score(X,y)
cross_val_score(logreg, X_train, y_train, cv = 3, scoring = 'accuracy')
```

```python
from sklearn.model_selection import cross_val_predict
y_train_pred = cross_val_predict(logreg, X_train, y_train, cv=3)

from sklearn.metrics import confusion_matrix 
confusion_matrix(y_train, y_train_pred)
```

```python
from sklearn.metrics import precision_score, recall_score
print("Precision Score =",precision_score(y_train.values, y_train_pred, pos_label = 'good'))
print("Recall Score =",recall_score(y_train.values, y_train_pred, pos_label = 'good'))
```

```python
from sklearn.metrics import f1_score
f1_score(y_train.values, y_train_pred, pos_label = 'good') 
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, y_train)
```

```python
y_pred1 = knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_true, y_pred1)
```

```python
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
%matplotlib inline
import sklearn.linear_model
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
df1 = pd.read_csv("C:/Users/Andrew Sundberg/Documents/DS 301/DS301-f19-master/DS301-f19-master/datasets/Brain-BodySize/brain-bodysize.csv")
df1.head(200)
```

```python
X = df1.Brain
Y = df1.Body
def best_fit(X, Y):

    xbar = sum(X)/len(X)
    ybar = sum(Y)/len(Y)
    n = len(X) # or len(Y)

    numer = sum([xi*yi for xi,yi in zip(X, Y)]) - n * xbar * ybar
    denum = sum([xi**2 for xi in X]) - n * xbar**2

    b = numer / denum
    a = ybar - b * xbar

    print('best fit line:\ny = {:.2f} + {:.2f}x'.format(a, b))

    return a, b

a, b = best_fit(X, Y)
plt.scatter(X, Y)
yfit = [a + b * xi for xi in X]
plt.plot(X, yfit)
```
